```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Tagged_data.xlsx')
```

```python
data.head()
```

```python
data=data.dropna()
```

```python
# checking the null values count and percentage in every column
count = data.isnull().sum().sort_values(ascending=False)
percentage = ((data.isnull().sum()/len(data))*100).sort_values(ascending=False)
missing_data = pd.concat([count,percentage],axis=1,keys=['count','percentage'])
missing_data
    
```

```python
data['feedback'] = data['textInput']
```

```python
data
```

```python
data.info()
```

```python
data[data['feedback'].apply(lambda x: isinstance(x,str))]
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
def cleanText(text):
    text=re.sub("[^a-zA-Z]", ' ', str(text))
    text=text.lower()
    text=text.split()
    ps = PorterStemmer()
    wl = WordNetLemmatizer()
    all_stopwords = stopwords.words('english')
#    all_stopwords.remove('not')
    not_stopwords = ['not','no']
    new_stopwords = [word for word in all_stopwords if not word in not_stopwords]
    text=[ps.stem(word) for word in text if not word in set(new_stopwords)]
    text=[wl.lemmatize(word) for word in text if not word in set(new_stopwords)]
    text=' '.join(text)
    return text
```

```python
data.head()
```

```python
clean = lambda x: cleanText(x)
data['Cleantext'] = pd.DataFrame(data.feedback.apply(clean))
data.head()
```

```python
data['Length']=data['textInput'].str.len()
data
```

## Model training

```python
from sklearn.model_selection import train_test_split

ind_var = data.Cleantext
dep_var = data.Sentiment

IV_train, IV_test, DV_train, DV_test = train_test_split(ind_var, dep_var, test_size = 0.3, random_state = 42)

print('IV_train :', len(IV_train))
print('IV_test  :', len(IV_test))
print('DV_train :', len(DV_train))
print('DV_test  :', len(DV_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

tvec = TfidfVectorizer()
clf2 = LogisticRegression(solver = "lbfgs")


from sklearn.pipeline import Pipeline

model = Pipeline([('vectorizer',tvec),('classifier',clf2)])

model.fit(IV_train, DV_train)


from sklearn.metrics import confusion_matrix

predictions = model.predict(IV_test)

confusion_matrix(predictions, DV_test)
```

```python
### Model Prediction

from sklearn.metrics import accuracy_score, precision_score, recall_score

print("Accuracy : ", accuracy_score(predictions, DV_test))
print("Precision : ", precision_score(predictions, DV_test, average = 'weighted'))
print("Recall : ", recall_score(predictions, DV_test, average = 'weighted'))
```
